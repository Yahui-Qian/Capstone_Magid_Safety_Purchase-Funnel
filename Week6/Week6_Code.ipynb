```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, roc_auc_score, average_precision_score, confusion_matrix, f1_score, precision_score
```

```python
# Set the Google Cloud project ID
project_id = "sublime-shift-464703-a2"
client = bigquery.Client(project=project_id)

# SQL query to load the BigQuery table
query = """
SELECT *
FROM `sublime-shift-464703-a2.funnel_output.week6`
"""

# Run the query and convert the result to a DataFrame
df = client.query(query).to_dataframe()
df.head()
```

```python
# Check all columns for missing values ​​(including NaN and None)
missing_summary = df.isnull().sum().to_frame(name='missing_count')
missing_summary['missing_percent'] = (missing_summary['missing_count'] / len(df)) * 100
missing_summary = missing_summary[missing_summary['missing_count'] > 0]  # Show only missing columns
missing_summary = missing_summary.sort_values(by='missing_count', ascending=False)

missing_summary
```

```python
import numpy as np
import pandas as pd

# ---------- 1) Basic Cleaning: fill missing values ----------
# Numeric
df['cart_value'] = df['cart_value'].fillna(0.0)
df['cart_quantity'] = df['cart_quantity'].fillna(0)

# String
df['item_name'] = df['item_name'].fillna('Unknown')
df['browser'] = df['browser'].fillna('Unknown')
df['medium'] = df['medium'].fillna('unknown')
df['source'] = df['source'].fillna('unknown')
df['coupon'] = df['coupon'].fillna('No Coupon')
df['promotion_name'] = df['promotion_name'].fillna('No Promotion')

# Date
if 'first_seen_date' in df.columns and 'session_date' in df.columns:
    df['first_seen_date'] = pd.to_datetime(df['first_seen_date'])
    df['session_date'] = pd.to_datetime(df['session_date'])
    df['first_seen_date'] = df['first_seen_date'].fillna(df['session_date'])

# ---------- 2) Feature Engineering: create new features ----------
# 2.1 Time on site (GA4 microseconds → seconds)
if 'session_first_event' in df.columns and 'session_last_event' in df.columns:
    df['time_on_site_sec'] = (df['session_last_event'] - df['session_first_event']) / 1e6
else:
    raise ValueError("Missing session_first_event / session_last_event (add these fields in SQL)")

# 2.2 Safe average item price
df['avg_item_price'] = np.where(df['cart_quantity'] > 0,
                                df['cart_value'] / df['cart_quantity'], 0.0)

# 2.3 New user flag (if SQL already has is_new_user, skip this step)
if 'first_seen_date' in df.columns and 'session_date' in df.columns:
    df['is_new_user'] = (df['session_date'] - df['first_seen_date']).dt.days.le(1).astype(int)
elif 'is_new_user' not in df.columns:
    df['is_new_user'] = 0

# 2.4 Promotion / coupon binary flags (handles "No Coupon/No Promotion/(not set)")
def _flag_used(x):
    s = ('' if x is None else str(x)).strip().lower()
    return 0 if s in {'(not set)','not set','none','no promotion','no coupon','unknown',''} else 1

df['used_coupon'] = df['coupon'].apply(_flag_used)
df['used_promo']  = df['promotion_name'].apply(_flag_used)

# 2.5 Time features (cyclical encoding)
if 'event_hour' in df.columns:
    df['event_hour'] = df['event_hour'].fillna(0).astype(int)
else:
    df['event_hour'] = 0
df['hour_sin'] = np.sin(2 * np.pi * df['event_hour'] / 24)
df['hour_cos'] = np.cos(2 * np.pi * df['event_hour'] / 24)

# 2.6 Funnel stage buckets (max_stage should be aggregated in SQL)
def stage_bucket(x):
    if pd.isna(x): return 'unknown'
    x = int(x)
    if x <= 1: return 'b1_view'
    if x == 2: return 'b2_cart'
    if x == 3: return 'b3_checkout'
    if x == 4: return 'b4_shipping'
    return 'b5_payment_or_more'
df['funnel_bucket'] = df['max_stage'].apply(stage_bucket)

# 2.7 Country grouping (reduce sparsity)
if 'geo_country' in df.columns:
    top = df['geo_country'].value_counts().index[:15]
    df['geo_country_grp'] = np.where(df['geo_country'].isin(top), df['geo_country'], 'Other')
else:
    df['geo_country_grp'] = 'Other'

# 2.8 Target variable
if 'converted_within_7d' not in df.columns:
    raise ValueError("Missing converted_within_7d (generate this label in SQL)")
y = df['converted_within_7d'].astype(int)

print("Cleaning + feature engineering done. Rows:", len(df))
```

```python
# ==================================
# 2) Feature column selection & preprocessing (One-Hot + Normalization)
# ==================================
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.impute import SimpleImputer

# Numerical features
num_cols = [
    'time_on_site_sec','cart_value','cart_quantity','avg_item_price',
    'is_new_user','event_hour','hour_sin','hour_cos','max_stage',
    'used_coupon','used_promo'
]
for c in num_cols:
    if c not in df.columns:
        df[c] = 0

# Classification features
cat_cols = []
for c in ['source','medium','device_type','browser','item_category','geo_country_grp','funnel_bucket']:
    if c in df.columns:
        cat_cols.append(c)

# Save the primary key & business columns for output
id_cols = [c for c in ['user_session_id','user_pseudo_id','session_id'] if c in df.columns]

# Preprocessor: numerical (missing → median, standardized), categorical (missing → "Unknown", one-hot)
numeric_tf = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler(with_mean=False))  # Sparse matrix friendly
])
categorical_tf = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore', sparse_output=False))
])

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_tf, num_cols),
        ('cat', categorical_tf, cat_cols)
    ],
    remainder='drop',
    sparse_threshold=0.3
)

# Split training/testing
X = df[num_cols + cat_cols]
X_train, X_test, y_train, y_test, idx_train, idx_test = train_test_split(
    X, y, df.index, test_size=0.2, random_state=42, stratify=y
)

print(f"Train: {len(X_train):,}, Test: {len(X_test):,}, Pos rate (train): {y_train.mean():.3f}")
```

```python
# ==================================
# 3) Model 1: Logistic Regression (baseline)
# ==================================
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.metrics import (
    classification_report, roc_auc_score, average_precision_score, confusion_matrix
)

logit_clf = Pipeline(steps=[
    ('prep', preprocessor),
    ('model', LogisticRegression(
        max_iter=1000,
        class_weight='balanced',  # Alleviate the imbalance of positive and negative samples
        solver='lbfgs',
        n_jobs=-1
    ))
])

logit_clf.fit(X_train, y_train)
logit_prob = logit_clf.predict_proba(X_test)[:, 1]
logit_pred = (logit_prob >= 0.5).astype(int)

print("=== Logistic Regression ===")
print("ROC AUC:", roc_auc_score(y_test, logit_prob))
print("PR AUC :", average_precision_score(y_test, logit_prob))
print(classification_report(y_test, logit_pred))
print("Confusion matrix:\n", confusion_matrix(y_test, logit_pred))
```

```python
# ==================================
# 4) Model 2: XGBoost (Main Model)
# ==================================
from xgboost import XGBClassifier

# Positive and negative sample ratio, set scale_pos_weight
pos_ratio = y_train.mean()
scale_pos_weight = (1 - pos_ratio) / max(pos_ratio, 1e-6)

xgb_clf = Pipeline(steps=[
    ('prep', preprocessor),
    ('model', XGBClassifier(
        n_estimators=600,
        learning_rate=0.05,
        max_depth=4,
        subsample=0.9,
        colsample_bytree=0.9,
        reg_lambda=2.0,
        min_child_weight=1.0,
        objective='binary:logistic',
        eval_metric='auc',
        tree_method='hist',
        scale_pos_weight=scale_pos_weight,
        random_state=42
    ))
])

xgb_clf.fit(X_train, y_train)
xgb_prob = xgb_clf.predict_proba(X_test)[:, 1]
xgb_pred = (xgb_prob >= 0.5).astype(int)

print("=== XGBoost ===")
print("ROC AUC:", roc_auc_score(y_test, xgb_prob))
print("PR AUC :", average_precision_score(y_test, xgb_prob))
print(classification_report(y_test, xgb_pred))
print("Confusion matrix:\n", confusion_matrix(y_test, xgb_pred))
```

## 1) Select threshold (two ways: maximum F1; or reach specified precision)

```python
import numpy as np
from sklearn.metrics import f1_score, precision_score, recall_score, roc_auc_score, average_precision_score, classification_report, confusion_matrix

# --------- Use XGB probability as the main model score ----------
y_score = xgb_prob # If using logistic regression, change to: y_score = logit_prob

# 1.1 Find the "maximum F1" threshold between [0.05, 0.95]
thr_grid = np.linspace(0.05, 0.95, 19)
f1s = [f1_score(y_test, (y_score >= t).astype(int)) for t in thr_grid]
best_idx = int(np.argmax(f1s))
best_thr_by_f1 = float(thr_grid[best_idx])
best_f1 = float(f1s[best_idx])

# 1.2 Find the minimum threshold that can achieve "Precision ≥ 0.30" (the target accuracy can be changed)
target_P = 0.30
thr_for_precision = None
for t in np.linspace(0.05, 0.95, 91):
    p = precision_score(y_test, (y_score >= t).astype(int))
    if p >= target_P:
        thr_for_precision = float(t)
        break

print(f"Best threshold by F1: {best_thr_by_f1:.2f} | F1={best_f1:.3f}")
print(f"Smallest threshold achieving Precision ≥ {target_P:.2f}: {thr_for_precision}")
```

## 3) Business Metrics: Precision@K, Lift Curve (Top 5/10/20% Focus)

```python
from sklearn.metrics import precision_score

def precision_at_k(y_true, y_score, k=0.10):
    n = max(int(len(y_score) * k), 1)
    idx = np.argsort(-y_score)[:n]  # Take the top K% from high to low probability
    return precision_score(y_true.iloc[idx] if hasattr(y_true, "iloc") else y_true[idx],
                           np.ones(n, dtype=int))

for k in (0.05, 0.10, 0.20):
    print(f"Precision@Top{int(k*100)}%:", round(precision_at_k(y_test, y_score, k), 3))

# --- Lift curve data (optional drawing) ---
import numpy as np
bins = 20
order = np.argsort(-y_score)
y_sorted = y_test.iloc[order] if hasattr(y_test, "iloc") else y_test[order]
chunk = len(y_sorted) // bins
lift = []
base_rate = y_test.mean()
for i in range(bins):
    seg = y_sorted[i*chunk:(i+1)*chunk]
    lift.append(seg.mean() / base_rate if base_rate > 0 else np.nan)

print("Lift by decile (1=Top5% … 5=Bottom5%):", [round(v, 2) for v in lift])


# ---- Draw the Lift curve ----
plt.figure(figsize=(6, 4))
plt.plot(range(1, bins + 1), lift, marker='o', color='b')
plt.axhline(1, color='gray', linestyle='--', linewidth=1)

plt.title('Lift Curve by Decile')
plt.xlabel('Decile (1 = Top 5%, 10 = Bottom 5%)')
plt.ylabel('Lift')
plt.xticks(range(1, bins + 1))
plt.grid(True)

# save image
plt.tight_layout()
plt.savefig("lift_curve.png", dpi=300)
plt.show()
```

```python
# === SHAP Interpretability (XGBoost) ===
import numpy as np
import pandas as pd
import scipy.sparse as sp
import shap
import matplotlib.pyplot as plt

# Get the trained components (your xgb_clf is a Pipeline containing a preprocessor)
xgb_model = xgb_clf.named_steps['model']
preproc   = xgb_clf.named_steps['prep']

# 1) Temporarily set the preprocessor output to numpy (to avoid pandas+sparse conflicts)
preproc.set_output(transform="default")

# 2) Encode X_test and make sure it is a dense matrix
X_test_enc = preproc.transform(X_test)
if sp.issparse(X_test_enc):
    X_test_enc = X_test_enc.toarray()
# 3) Get the encoded feature name
try:
    feature_names = preproc.get_feature_names_out()
except Exception:
    feature_names = [f"f{i}" for i in range(X_test_enc.shape[1])]
# 4) Calculate SHAP (xgboost tree model)
explainer    = shap.TreeExplainer(xgb_model)
shap_values  = explainer(X_test_enc)          # Explanation Object
# Let beeswarm display our own column names
shap_values.feature_names = feature_names

# ========= A. Importance of Original Field Granularity (One-Hot Aggregation) =========
# 4.1 First, take the mean(|SHAP|) value of each encoded column.
mean_abs_by_encoded = np.abs(shap_values.values).mean(axis=0)
encoded_importance  = pd.Series(mean_abs_by_encoded, index=feature_names, name="mean_abs_shap")

# 4.2 Mapping "encoded column name -> original field name"
def to_raw_feature(name: str) -> str:
    ""
    Compatible with ColumnTransformer naming:
      - Value: num__time_on_site_sec -> time_on_site_sec
      - Category: cat__source_google -> source
      - Other: Return directly
    ""
    parts = name.split("__", 1)
    if len(parts) == 2:
        block, rest = parts
        # One-Hot expansion columns such as source_google / browser_chrome, etc.
        if block.startswith("cat"):
            return rest.split("_", 1)[0]
        else:
            return rest
    return name

raw_index = [to_raw_feature(c) for c in encoded_importance.index]
raw_importance = (
    encoded_importance
      .groupby(raw_index)         # Aggregate by original field (sum the one-hot subcolumns)
      .sum()
      .sort_values(ascending=False)
)

# 4.3 Display & Save Top-K (Original Field Granularity)
TOP_K = 20
top_raw = raw_importance.head(TOP_K)
display(top_raw.to_frame().rename_axis("raw_feature"))

plt.figure(figsize=(7, 7))
top_raw.iloc[::-1].plot(kind="barh", legend=False)
plt.title(f"SHAP Importance by Raw Feature (Top {TOP_K})")
plt.xlabel("mean(|SHAP value|)")
plt.tight_layout()
plt.savefig("shap_importance_raw_topK.png", dpi=300)
plt.show()

print("已保存：shap_importance_raw_topK.png")

# ========= B. Standard beeswarm (encoded column), see the specific value mode =========
plt.figure(figsize=(8, 8))
shap.plots.beeswarm(shap_values, max_display=20, show=True)
plt.savefig("shap_beeswarm_encoded_top20.png", dpi=300, bbox_inches="tight")
print("已保存：shap_beeswarm_encoded_top20.png")

# (Optional) Top 20 bar chart of encoded columns (no aggregation)
plt.figure(figsize=(7, 7))
(pd.Series(mean_abs_by_encoded, index=feature_names)
     .sort_values(ascending=False)
     .head(20)
     .iloc[::-1]
     .plot(kind="barh", legend=False)
)
plt.title("SHAP Importance by Encoded Feature (Top 20)")
plt.xlabel("mean(|SHAP value|)")
plt.tight_layout()
plt.savefig("shap_importance_encoded_top20.png", dpi=300)
plt.show()

# 5) Restore output to pandas (if you are used to using pandas in subsequent processes)
preproc.set_output(transform="pandas")

```
