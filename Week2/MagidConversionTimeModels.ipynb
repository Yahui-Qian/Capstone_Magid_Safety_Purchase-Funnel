```python
import pandas as pd
from pycaret.classification import ClassificationExperiment, create_model, tune_model
```

```python
data = pd.read_csv('/content/drive/MyDrive/final_input_downsampled_0801.csv')
data = data.drop(columns=['user_pseudo_id', 'session_id'])
data.head()
```

```python
s = ClassificationExperiment()
```

```python
s.setup(data, target = 'time_to_conversion', session_id = 123)

# model training and selection
best = s.compare_models()
```

```python
# knn
s.setup(data, target = 'time_to_conversion', session_id = 123,
        normalize=False, imputation_type=None)

knn_grid = {'n_neighbors': [5],
            'weights': ['uniform', 'distance'],
            'algorithm': ['auto', 'ball_tree', 'kd_tree', 'brute']}
knn = s.create_model('knn')
tuned_knn = s.tune_model(knn, custom_grid = knn_grid, optimize = 'F1')
print(tuned_knn)
```
