```python
import pandas as pd
```

```python
data = pd.read_csv('session_sequences.csv')
```

```python
data.shape
```

```python
# count event_name and percentage for initial session
initial_session = data[data['session_label'] == 'initial']

initial_counts = initial_session['event_name'].value_counts()
initial_percentage = initial_counts / len(initial_session) * 100

initial_df = pd.DataFrame({
    'event_name': initial_counts.index,
    'count': initial_counts.values,
    'percentage': initial_percentage.values
})
```

```python
initial_df[:30]
```

```python
initial_df[30:]
```

```python
# returned sessions
returned_session = data[data['session_label'] == 'returned']
returned_counts = returned_session['event_name'].value_counts()
returned_percentage = returned_counts / len(returned_session) * 100
returned_df = pd.DataFrame({
    'event_name': returned_counts.index,
    'count': returned_counts.values,
    'percentage': returned_percentage.values
})
```
